```python
import pandas as pd
```

```python
data = pd.DataFrame({ 'gender' : ['Male', 'Female', 'Male', 'Female', 'Female'],
                      'class' : ['A','B','C','D','A'],
                      'city' : ['Rio de Janeiro','Gurugram','Santos','Delhi','Gurugram'] ,
                      'classification': ['Violent', 'Quiet', 'Very Quiet', 'Medium', 'Quiet']})
data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
oh = OneHotEncoder(sparse_output=False).set_output(transform='pandas')
```

```python
oh.fit(data[['city']])
oh.categories_
```

```python
onehoy_encoded = oh.fit_transform(data[['city']])
```

```python
df = pd.concat([data,onehoy_encoded],axis =1).drop(columns=['city'])
df.head()
```

```python
from sklearn.preprocessing import LabelBinarizer

lb = LabelBinarizer()
lb.fit(data['city'])
lb.classes_
```

```python
cols = lb.fit_transform(data['city'])
temp_df = pd.DataFrame(cols)
temp_df
```

```python
df2 = data.join(temp_df)
df2.head()
```
